```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_excel("data.xlsx", sheet_name="Sheet1")

# Create scatter plot
fig, ax = plt.subplots()
ax.scatter(df["波長 [nm]"], df["電圧 [VDC]"], marker='o', color='black')

# Set labels
plt.xlabel("wavelength [nm]", fontsize=14)
plt.ylabel("voltage [VDC]", fontsize=14)

# Add minor ticks inside the plot
ax.tick_params(which='both', direction='in')

# Save plot as PDF file
fig.savefig("scatter_plot.pdf", format="pdf")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Excelファイルからデータを読み込む
df = pd.read_excel('data.xlsx', sheet_name='Sheet3')

# データから2次関数を推定
x = df['回転角度']
y = df['波長']
a, b, c = np.polyfit(x, y, 2)

# 推定された係数を使用して描画データを生成
x_new = np.linspace(min(x), max(x), 100)
y_new = a * x_new ** 2 + b * x_new + c

# グラフ作成
fig, ax = plt.subplots()
ax.plot(x, y, 'o')
ax.plot(x_new, y_new, '-', label='fitted curve', color='blue')
plt.ylim(0, 800)
plt.xlim(212, 220)

# 補助目盛を内側に表示
ax.tick_params(which='both', direction='in', top=True, right=True)

# 方程式をグラフに表示
eqn_text = r'$\lambda = {:.2f}\theta^2 + {:.2f}\theta + {:.2f}$'.format(a, b, c)
eqn_coord = (max(x) + min(x)) / 2, y_new[np.argmin((y_new - max(y_new) / 2) ** 2)] - 50
ax.text(eqn_coord[0], eqn_coord[1], eqn_text, fontsize=10, ha='center')
ax.set_xlabel('Angle of rotation $θ$ [°]', fontsize=14)
ax.set_ylabel('wavelegth $\lambda$ [nm]', fontsize=14)
ax.legend()

plt.show()
fig.savefig("fitted_curve.pdf", format="pdf")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Excelファイルを読み込む
df = pd.read_excel('data.xlsx', sheet_name='Sheet4')

# 波長(x)と電圧(y)の列を取得
x_values = df['波長']
y_values = df['電圧']

# 棒グラフを描画
plt.bar(x_values, y_values)

# グラフのタイトルや軸ラベルを設定
plt.xlabel('wavelength [nm]', fontsize=14)
plt.ylabel('voltage [mVDC]', fontsize=14)

# ファイルを保存
plt.savefig('output.pdf')

# Set tick marks to face inward
plt.rcParams["xtick.direction"] = "in"
plt.rcParams["ytick.direction"] = "in"
plt.rcParams['xtick.top'] = True
plt.rcParams['ytick.right'] = True

# 表示
plt.show()
```
