```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Define data as an array
x = np.array([4.98, 5.37, 6.21, 7.02])
y = np.array([0.40, 0.56, 0.78, 1.02])

# Estimate linear functions
a, b = np.polyfit(x, y, 1)

# Generate drawing data using estimated coefficients
x_new = np.linspace(0, max(x), 100)
y_new = a * x_new + b

# Create graphs
fig, ax = plt.subplots()
ax.plot(x, y, 'o')
ax.plot(x_new, y_new, '-', label='fitting line', color='blue')

# Graph setup
plt.xlabel('f (Frequency) [$10^{14}$Hz]')
plt.ylabel('V (Blocking Voltage) [V]')
plt.xlim(0, max(x) + 1)  # x軸の表示範囲を0から最大値+1に設定
plt.legend()

# Extend and draw a straight line
x_extension = np.linspace(min(x), 0, 100)
y_extension = a * x_extension + b
ax.plot(x_extension, y_extension, '--', color='gray', alpha=0.5)

plt.tick_params(direction='in')
# グラフを表示
plt.show()

fig.savefig('figs/fitting.pdf')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Define data as an array
x = np.array([4.98, 5.37, 6.21, 7.02])
y = np.array([0.40, 0.56, 0.78, 1.02])

# Estimate linear functions
a, b = np.polyfit(x, y, 1)

# Generate drawing data using estimated coefficients
x_new = np.linspace(0, max(x), 100)
y_new = a * x_new + b

# Define additional data arrays
x2 = np.array([4.98, 5.37, 6.21, 7.02])
y2 = np.array([0.2865,0.3961,0.542,0.6604])

x3 = np.array([4.98, 5.37, 6.21, 7.02])
y3 = np.array([0.3682	,0.4187	,0.7034	,0.9061])

# Estimate linear functions for additional data
a2, b2 = np.polyfit(x2, y2, 1)
a3, b3 = np.polyfit(x3, y3, 1)

# Generate drawing data using estimated coefficients for additional data
x2_new = np.linspace(0, max(x2), 100)
y2_new = a2 * x2_new + b2

x3_new = np.linspace(0, max(x3), 100)
y3_new = a3 * x3_new + b3

plt.rcParams.update({'font.size': 8})

# Create graphs
fig, ax = plt.subplots()
ax.plot(x, y, 'o', label='Calculated from a current measurement of 0 [A]')
ax.plot(x_new, y_new, '-', label='$V = 2.959\\times10^{-15}f - 1.0544$', color='blue')
ax.plot(x2, y2, 'o', label='Calculated from the linear part of data')
ax.plot(x2_new, y2_new, '-', label='$V = 1.79\\times10^{-15}f - 0.5841$', color='orange')
ax.plot(x3, y3, 'o', label='Calculated from the curve part of data')
ax.plot(x3_new, y3_new, '-', label='$V = 2.761\\times10^{-15}f - 1.0288$', color='green')

plt.tick_params(direction='in')
# Graph setup
plt.xlabel('f (Frequency) [$10^{14}$Hz]')
plt.ylabel('V (Blocking Voltage) [V]')
plt.xlim(0, max(x3) + 1)  # Set x-axis limit
plt.legend()

# Extend and draw a straight line for all data points
x_extension = np.linspace(min(x), 0, 100)
y_extension = a * x_extension + b
ax.plot(x_extension, y_extension, '', color='blue', alpha=0.5)

x2_extension = np.linspace(min(x2), 0, 100)
y2_extension = a2 * x2_extension + b2
ax.plot(x2_extension, y2_extension, '', color='orange', alpha=0.5)

x3_extension = np.linspace(min(x3), 0, 100)
y3_extension = a3 * x3_extension + b3
ax.plot(x3_extension, y3_extension, '', color='green', alpha=0.5)
plt.tick_params(direction='in')
# Display the plo
plt.xlim(0, 12)  # Set x-axis range from 4 to 8
plt.ylim(-1.2, 1.2) # Set y-axis range from 0 to 1.5

plt.show()

fig.savefig('figs/fitting.pdf')
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['text.usetex'] = True # TeX記法を使えるようにする

def f(x):
    return np.exp(1/x) / (1 + np.exp(1/x))**2 / x**2

x = np.linspace(0, 5, 500) # グラフに描画するxの値

plt.plot(x, f(x))
plt.legend(['f(x)'], loc='upper right',fontsize=12) # 凡例を右上に表示
plt.xlim([0, 5]) # x軸の範囲を指定
plt.ylim([0, 0.5]) # y軸の範囲を指定
plt.axhline(y=0, color='black', lw=1) # y軸を0に合わせる
plt.axvline(x=0, color='black', lw=1) # x軸を0に合わせる
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title(r'$f(x) = \exp^{1/x}/{(1 + \exp^{1/x})^2 x^2}$') # タイトル
plt.tick_params(direction='in') # メモリ線を内側にする
plt.savefig('figs/graph.pdf')
plt.show()
```
