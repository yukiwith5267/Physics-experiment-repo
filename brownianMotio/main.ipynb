```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the data from the Excel file
df = pd.read_excel('data.xlsx', sheet_name='1_traces')

# Create the scatter plot for all data
plt.scatter(df['t'], df['MSD'])

# Select data up to the 13th entry for linear regression
df_subset = df.iloc[:13]

# Perform linear regression on the subset
slope, intercept = np.polyfit(df_subset['t'], df_subset['MSD'], 1)

# Create the line of best fit
line = slope * df_subset['t'] + intercept

# Plot the approximate straight line on the subset
plt.plot(df_subset['t'], line, 'r-')

# Set the title and labels for the plot
plt.xlabel('t[s]')
plt.ylabel(r'$\mathrm{MSD [μm^2/s]}$')
plt.tick_params(direction='in')

# Extend the line of best fit
extended_line = slope * df['t'] + intercept

# Plot the extended line
plt.plot(df['t'], extended_line, 'r-')

# Set the y-axis range
plt.ylim(0, 4)

# Calculate the equation of the line
equation = f'y = {slope:.3f}x + {intercept:.2f}'

# Annotate the equation on the plot
plt.annotate(equation, (df_subset['t'].max(), line.max()), xytext=(-50, 30),
       textcoords='offset points', arrowprops=dict(arrowstyle='->'))


plt.savefig('fig/1_D.pdf')
```
