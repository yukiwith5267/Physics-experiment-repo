```python
import numpy as np
import matplotlib.pyplot as plt

# 定数の定義
C = 10e-6  # 10μF
R = 10    # 10Ω

# fの範囲を定義
f_smooth = np.logspace(2, 5, 1000)  # 100から100000Hzまでの範囲で計算 (対数スケール)

# 第1のG(f)の計算
G1_smooth = (2 * np.pi * f_smooth * C)**2 * R / (1 + (2 * np.pi * f_smooth * C * R)**2)

# 第2のG(f)の計算
G2_smooth = (2 * np.pi * f_smooth * C) / (1 + (2 * np.pi * f_smooth * C * R)**2)

# グラフの描画
fig, ax1 = plt.subplots(figsize=(10, 6))

# 左のy軸にG1をプロット
ax1.set_xlabel('Frequency (Hz)', fontsize=16)
ax1.set_ylabel('G [S]', color='red', fontsize=16)
ax1.plot(f_smooth, G1_smooth, label='G', color='r')
ax1.tick_params(axis='y', labelcolor='r', direction='in',labelsize=16)
ax1.tick_params(axis='x', direction='in', labelsize=16)
ax1.set_xscale('log')
ax1.grid(False)  # Remove grid lines for ax1

# 右のy軸にG2をプロット
ax2 = ax1.twinx()
ax2.set_ylabel('B [S]', color='blue', fontsize=16)
ax2.plot(f_smooth, G2_smooth, label='B', color='b')

ax2.tick_params(axis='y', labelcolor='b', direction='in', labelsize=16)
ax2.grid(False)  # Remove grid lines for ax2

# GとBの判例を追加
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
lines = lines1 + lines2
labels = labels1 + labels2
ax1.legend(lines, labels, loc='upper right', fontsize=14)

fig.savefig('fig/rc_admittance.pdf')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 定数の値
C = 10e-6  # 10 μFをファラッドに変換
R = 100    # 100 Ω

# 周波数の範囲を設定
f = np.logspace(0, 5, 1000)  # 1 Hzから100 kHzまでの周波数範囲を作成

# C' と C'' の関数を計算
C_prime = C / (1 + (2 * np.pi * f * C * R) ** 2)
C_double_prime = (2 * np.pi * f * C ** 2 * R) / (1 + (2 * np.pi * f * C * R) ** 2)

# グラフを描画
fig, ax1 = plt.subplots(figsize=(10, 6))

# 第1のy軸 (左側) にC'をプロット
ax1.semilogx(f, C_prime * 1e6, label="C'", color='b')  # μFに変換してプロット
ax1.set_xlabel('Frequency (Hz)', fontsize=16)
ax1.set_ylabel("C' [μF]", color='r', fontsize=16)
ax1.tick_params(axis='y', labelcolor='r', direction='in', labelsize=16)
ax1.tick_params(axis='x', labelsize=16, direction='in')
# 第2のy軸 (右側) にC''をプロット
ax2 = ax1.twinx()
ax2.semilogx(f, C_double_prime * 1e6, label="C''", color='r')  # μFに変換してプロット
ax2.set_ylabel("C'' [μF]", color='b', fontsize=16)
ax2.tick_params(axis='y', labelcolor='b', direction='in', labelsize=16)

lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
lines = lines1 + lines2
labels = labels1 + labels2
ax1.legend(lines, labels, loc='upper right', fontsize=14)

plt.show()

fig.savefig('fig/rc_capacita.pdf')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

# Load data from Excel file for the first scatter plot
df1 = pd.read_excel('data_2.xlsx', sheet_name='55')

# Define the model function with the updated parameter names for the first scatter plot
def epsilon_prime(f, epsilon_0, epsilon_infinity, tau):
    return (epsilon_0 - epsilon_infinity) / (1 + (2 * np.pi * f * tau)**2) + epsilon_infinity

# Fit the model to the data for the first scatter plot
params1, covariance1 = curve_fit(epsilon_prime, df1['f'], df1['ε\''])

# Extract the optimized parameters for the first scatter plot
epsilon_0_opt1, epsilon_infinity_opt1, tau_opt1 = params1

# Calculate f_r
f_r = 1 / (2 * np.pi * tau_opt1)

# Output f_r
print("f_r = ", f_r)

# Load data from Excel file for the second scatter plot
df2 = pd.read_excel('data_2.xlsx', sheet_name='55')

# Extract data columns for the second scatter plot
x_data2 = df2['f']
y_data2 = df2['ε\'\'']

# Define the epsilon approximation function for the second scatter plot
def epsilon_approx(f, epsilon_0, epsilon_infinity, tau):
    return ((epsilon_0 - epsilon_infinity) * 2 * np.pi * f * tau) / (1 + (2 * np.pi * f * tau) ** 2) + epsilon_infinity

# Initial guess for parameters for the second scatter plot
initial_guess2 = (0, 0, 0.000001)

# Perform the curve fitting for the second scatter plot
optimal_params2, _ = curve_fit(epsilon_approx, x_data2, y_data2, p0=initial_guess2)

# Extract the optimized parameters for the second scatter plot
epsilon_0_opt2, epsilon_infinity_opt2, tau_opt2 = optimal_params2

# Create the fitted curve using the optimized parameters for the second scatter plot
fitted_curve2 = epsilon_approx(x_data2, epsilon_0_opt2, epsilon_infinity_opt2, tau_opt2)

# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Plot the first scatter plot in the first subplot
axes[0].scatter(df1['f'], df1['ε\''], label='Data')
axes[0].plot(df1['f'], epsilon_prime(df1['f'], epsilon_0_opt1, epsilon_infinity_opt1, tau_opt1), label='Fitted Curve', color='red')
axes[0].set_xlabel('f')
axes[0].set_ylabel('ε\'')
axes[0].set_xscale('log')
axes[0].legend()
axes[0].text(0.7, 0.8, r"$\varepsilon_0 = {:.7g}$".format(epsilon_0_opt1), transform=axes[0].transAxes)
axes[0].text(0.7, 0.75, r"$\varepsilon_\infty = {:.7g}$".format(epsilon_infinity_opt1), transform=axes[0].transAxes)
axes[0].text(0.7, 0.7, r"$\tau = {:.7g}$".format(tau_opt1), transform=axes[0].transAxes)

# Plot the second scatter plot in the second subplot
axes[1].scatter(x_data2, y_data2, label='Data')
axes[1].plot(x_data2, fitted_curve2, label='Fitted Curve', color='red')
axes[1].set_xlabel('f')
axes[1].set_ylabel('ε\'\'')
axes[1].set_xscale('log')
axes[1].legend()

# Adjust spacing between subplots
plt.tight_layout()

# Set ticks position to inside for both subplots
for ax in axes:
    ax.tick_params(axis='both', which='both', direction='in', bottom=True, top=True, left=True, right=True)

plt.savefig('fig/fit_55.pdf')
plt.show()
```
